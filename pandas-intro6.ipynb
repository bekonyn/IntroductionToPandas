## Apply Function to Pandas Series and DataFrames

```python
import pandas as pd
train=pd.read_csv("../input/titanic_train.csv")
train.head()
```

```python
#lets create dummy variable for sex-there is much more ez,faster way to create dummy variable check the prev notes for that
train["Sex_num"]=train.Sex.map({"female":0,"male":1})
#map can be used for creating new col with old cols catg data's made up indices.
```

```python
train.loc[0:4,["Sex","Sex_num"]]
```

```python
#we want to create a col that has name's col values string count lets do it.apply method applies length to all all of them
#this apply function is similar to R's apply functions apply,sapply,tapply,lapply,vapply etc
train["Name_Length"]=train.Name.apply(len)
```

```python
train.loc[0:4,["Name","Name_Length"]]
```

```python
#we ll use apply function-similar to apply f() in R.
#lets round the values in Fare col. Use ceil function to be more exact.
import numpy as np
train["Fare_Ceil"]=train.Fare.apply(np.ceil)
#apply applies the function/method u specify to the data/col u specify
```

```python
train.loc[0:4,["Fare","Fare_Ceil"]]
```

```python
#we want to split our name cols from their commas
train.Name.str.split(",").head()
#our data converted from string to list of strings. All of it is a series ofc.its series of python string list.
```

```python
#we want first part of this string which is Braund,Cumings,Heikkinen etc.lets create a function for it
def get_element(my_list,position):
    return my_list[position]
#if a pass a list and say pos=0 it returns first value of that list which is Braund,Cumings,Heikkinen for our ex.
```

```python
train.Name.str.split(",").apply(get_element,position=0).head()
#take train.Name.str.split(",") this Series apply get_element f() to every element and pass the pos=0
#but sth this simple creating yr own f() is unnecessary so lets use lambda function
```

```python
#x is every row of that train.Name.str.split(",") series.lambda functions used a lot w apply functions
train.Name.str.split(",").apply(lambda x: x[0]).head()
```

```python
drinks=pd.read_csv("../input/drinks.csv")
drinks.head()
```

```python
#here is our dataset(dataframe) we ll be working.sliced.LETS Use apply w a dataframe not a series this time.
drinks.loc[:,"beer_servings":"wine_servings"]
```

```python
#gets max value for each col.when we are using drop f() axis=1 is for col. but here axis=0 is for col you have to remember it
drinks.loc[:,"beer_servings":"wine_servings"].apply(max,axis=0)
```

```python
#if u care about max value of each row without its number just their colname
drinks.loc[:,"beer_servings":"wine_servings"].apply(np.argmax,axis=1)
#for row 1, max value's col in that row is spirit_servings. etc
```

```python
#we covered apply as series and df method now its time for applymap which is dataframe method.
#applies the function to whole dataframe not just a selected cols.
```

```python
#turn every data to float dtype
drinks.loc[:,"beer_servings":"wine_servings"].applymap(float)
```

## Using MultiIndex

```python
import pandas as pd
```

```python
stocks=pd.read_csv("../input/stocks.csv")
stocks.head()
```

```python
#date-closing price from that date-volume-stock symbol(csco=cisco,aapl=apple,msft=microsoft)
stocks.index
#its range index,regular index not a multi index
```

```python
stocks.groupby("Symbol").Close.mean()
```

```python
#here is another use of groupby,little unnecessary for this ex but u can see the capabilities
#group1=symbol subgroup=date 
ser=stocks.groupby(["Symbol","Date"]).Close.mean()
ser
```

```python
#ser dataset has multindex now.(multi-index=mi)mi adds extra dims.series usually 1d but series with mi is 2d,kind of..
#df w regular index is 2d with multi index its 3d,kind of.
ser.index
```

```python
#when we unstacked the ser (which is series) it becomes dataframe.our "kind of 2d" example comes from here.
df=ser.unstack()
df
```

```python
#if we want to get aapl data only.
ser.loc["AAPL"]
#or df.loc["AAPL"] just works fine
#df.loc["AAPL",:]
```

```python
#if we only want to know 2016-10-03 data about aapl
ser.loc["AAPL","2016-10-03"]
```

```python
#if u want to get all "2016-10-03"
ser.loc[:,"2016-10-03"]
```

```python
#we can use groupby to create more than 1 grouping criteria then unstack it which makes it a dataframe.now we can use our 
#df tools on it such as loc etc.
```

```python
#create a category column during file reading
drinks=pd.read_csv("../input/drinks.csv")
drinks.head()
```

```python
#when u have repeating string types as object(string) it may be good idea to convert them into categorical data.
drinks.dtypes
```

```python
#old way of changing your col from object to category(categorical data)
drinks["continent"]=drinks.continent.astype("category")
drinks.dtypes
#it still works but new format is much faster and its occured on data importing phase.
```

```python
drinks=pd.read_csv("../input/drinks.csv",dtype={"continent":"category"})
drinks.dtypes
#faster.creating dictionary to assing date types for cols.u can convert multiple cols at once but there is much faster way...
```

```python
#convert data type of multiple cols at once:
drinks=pd.read_csv("../input/drinks.csv")
drinks=drinks.astype({"beer_servings":"float","spirit_servings":"float"})
drinks.dtypes
```

```python
#apply multiple aggregations on a series or DATAframe
```

```python
#here is example of a single aggregation f() after a groupby
drinks.groupby("continent").beer_servings.mean()
```

```python
#multiple aggregation functions can be applied simultaneously using agg method to grouped data.
drinks.groupby("continent").beer_servings.agg(["mean","max","min"])
```

```python
#apply same aggregation to a Series.Returns a series.
drinks.beer_servings.agg(["mean","max","min"])
```

```python
#could be applied to all cols of dataframe.ofc for non numerical data it returns alphabetical result.
drinks.agg(["mean","max","min"])
```

```python
#less flexible approach.
drinks.describe()
```

```python
#DROP ACCEPTS "INDEX" AND "COLUMNS" KEYWORDS
ufo=pd.read_csv("../input/ufo.csv")
ufo.head()
```

```python
#in older version what u should have done to drop rows: specify labels and axis. labels(indexes) are row identifiers.can be string or numbers.
ufo.drop([0,1],axis=0).head()
#u specify the rows u want to omit/drop then axis=0 to specify that u want row drop.not a col drop.
```

```python
#u can use index rather than 0.set inplace=True if u want to make it permanent
ufo.drop([0,1],axis="index").head()
```

```python
#alternative approach.by default axis=0 so u dont have to specify it.if u specify indexes pandas consider them as rows
ufo.drop(index=[0,1]).head()
```

```python
#dropping cols
ufo.drop(["City","State"],axis=1).head()
```

```python
#this time panda cant auto recognize it. u have to specify it via columns arg.
ufo.drop(columns=["City","State"]).head()
```

```python
#RENAME AND REINDEX ACCEPT "AXIS" KEYWORD.
ufo.head()
```

```python
#rename and reindex are opposite.
#older versions approach  for renaming:
ufo.rename(columns={"City":"CITY","State":"STATE"}).head()
```

```python
#much newer approach
ufo.rename({"City":"CITY","State":"STATE"},axis=1).head()
#i get use to older format but if you shift tab inside parantheses u can see the args and can remember the axis mentality
```

```python
ufo.rename({"City":"CITY","State":"STATE"},axis="columns").head()
```

```python
#bonus tip
#if u want to set all your colnames to be upper...
ufo.rename(str.upper,axis=1).head()
#or axis="columns"
```

```python
#lets create a small df
df=pd.DataFrame({"ID":[100,101,102,103],"quality":["good","very good","good","excellent"]})
df
```

```python
#now convert it to a categorical data.This is  an older way,deprecated.
df.quality.astype("category",categories=["good","very good","excellent"],ordered=True)
```

```python
#much newer approach
from pandas.api.types import CategoricalDtype
quality_cat=CategoricalDtype(["good","very good","excellent"],ordered=True)
df["quality"]=df.quality.astype(quality_cat)
df.quality
#older way much ez and readable i cant get the reason behind the change.However newer approaches should be implemented.
```
