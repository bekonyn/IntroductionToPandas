## Filtering Pandas DataFrame

```python
import pandas as pd
```

```python
movies=pd.read_csv("../input/imdb_1000.csv")
```

```python
movies.head()
```

```python
#shape is df attribute, no parantheses is needed.
movies.shape
```

```python
#lets create a list and using loops to store the mov duration criteria for 979 movies.
booleans=[]
for i in movies.duration:
    if i>=200:
        booleans.append(True)
    else:
        booleans.append(False)
        
booleans[0:5]
```

```python
#to get element count in list...
len(booleans)
```

```python
#lets convert list to series
is_long=pd.Series(booleans)
#now it becomes a list and we can use head method just like in df
is_long.head()
```

```python
movies[is_long]
```

```python
#much more faster filtering(no loop is necessary) if u didnt import the pandas then just by using built in python..
#python ll give u one true/false value as a result.However pandas understand what u try to do.
is_long=movies.duration>=200
movies[is_long]
#movies[movies.duration>=200] ll work out and its much more general format to use.
```

```python
movies[movies.duration>=200]["genre"]
#or .genre but sts both of them could result some anormalities so we could use .loc method.
#loc helps u out selecting rows/cols by label. so u say to loc [row_i_want,col_i_want]
```

```python
movies.loc[movies.duration>=200,"genre"]
```

## Multiple Filtering

```python
import pandas as pd
```

```python
movies=pd.read_csv("../input/imdb_1000.csv")
```

```python
#one criteria is easy.what about 2 or more?
movies[movies.duration>=200]
```

```python
#approach 1
movies[(movies.duration>=200)& (movies.genre=="Drama")]
```

```python
#just running the boolean part.
(movies.duration>=200)& (movies.genre=="Drama").head()
```

```python
#bonus tips
#what if u want to filter for MULTIPLE or conditions. movie[(criteria1)| (criteria2) | (criteria3)] will work out just fine
#but it is too cumbersome, we need another way. "isin" ll help us out.its a series method.
movies.genre.isin(["Drama","Action","Adventure"])
```

```python
movies[movies.genre.isin(["Drama","Action","Adventure"])].head()
```

## Tips/Tricks about Pandas-1

```python
import pandas as pd
```

```python
ufo=pd.read_csv("../input/ufo.csv",
               usecols=["City","State"])
ufo.head()
#u can use col indexes as well, rather than colnames. usecols[0,3]. it ll work out just fine.
```

```python
#what if u want just few rows to import 
ufo=pd.read_csv("../input/ufo.csv",
               nrows=3)
ufo.head()
```

```python
#how do df and series work with regards to selecting indiviual entries and iteration.
```

```python
#iterating pandas series
for i in ufo.City:
    print(i)
```

```python
#iterating through dataframe(df)
for index,row in ufo.iterrows():
    print(index,row.City,row.State)
```

```python
#u dont have to use index it doesnt have any meaning to pandas, but u have to specify sth so it takes the index otherwise
#iteration wont work.pandas doesnt auto assign indexes to your iteration.
for n,row in ufo.iterrows():
    print(n,row.City,row.State)
```

## String Handling Methods in Pandas

```python
import pandas as pd
```

```python
#for series. its little different to call these functions for dataframe.
"hello".upper()
```

```python
orders=pd.read_table("../input/chipotle.tsv")
```

```python
orders.head()
```

```python
#for dataframe, u need to add string method.
orders.item_name.str.upper()
```

```python
#another string method use to check certain words in col data.
orders[orders.item_name.str.contains("Chicken")]
```

```python
#ofc it returns booleans if we seperate it and run it by itself
orders.item_name.str.contains("Chicken")
```

```python
#for more string methods-pandas api reference then u need to search for "string handling"
#http://pandas.pydata.org/pandas-docs/stable/reference/index.html
```

```python
#bonus tips
#we want to remove brackets "[]" that col data's have.
#orders.choice_description.str.replace("[","") returns a series(vectors in R) so u can chained str.replace again 
orders.choice_description.str.replace("[","").str.replace("]","")
```

```python
#bonus tips2
#get same result using regular expression \ for ex
orders.choice_description.str.replace("[\[\]]","")
```

## Convert Data Type of Pandas Series

```python
import pandas as pd
```

```python
drinks=pd.read_csv("../input/drinks.csv")
```

```python
drinks.dtypes
```

```python
drinks["beer_servings"]=drinks.beer_servings.astype(float)
```

```python
drinks.dtypes
```

```python
#how do we define imported data dtypes while importing them
drinks=pd.read_csv("../input/drinks.csv",
                  dtype={"beer_servings":float})
drinks.dtypes
```

```python
orders=pd.read_table("../input/chipotle.tsv")
```

```python
orders.head()
```

```python
orders.dtypes
```

```python
#even i remove $ sign its still a string so i cant do anything math related with it.let's solve it.
#orders.item_price.str.replace("$","").mean() doesnt work. we have categorical data however our model needs sth numerical.
```

```python
#this is a series so i can use astype(float) with it.
orders.item_price.str.replace("$","").astype(float).mean()
```

```python
#bonus tips.....we want to convert true/false values to 0-1 for machine learning models.
orders.item_name.str.contains("Chicken")
```

```python
#here is the 0-1 format for ml models.
orders.item_name.str.contains("Chicken").astype(int).head()
```

```python
#just to remember how can we use str.contains as a filter to dataframe.
orders[orders.item_name.str.contains("Chicken")].head()
```
