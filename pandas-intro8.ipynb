## Creating a New Col or Multi Cols Based on Pre Existing Cols

```python
#basically-how can i create new col based on pre existing col in pandas?
#u rarely need to use for loops in pandas.
```

```python
import pandas as pd
drinks = pd.read_csv('../input/drinks.csv')
movies = pd.read_csv('../input/imdb_1000.csv')
train = pd.read_csv('../input/titanic_train.csv')
ufo = pd.read_csv('../input/ufo.csv', parse_dates=['Time'])
```

```python
train.head()
```

```python
#lets create a col that says "young" when age is below 30 and also male etc
(train.Sex=="male") & (train.Age<30)
```

```python
#we use mapping.if we want to change the inputs of our data we use mapping(map() function)
train["young_male"]=(train.Sex=="male") & (train.Age<30).map({True:"Yes",False:"No"})
train.head()
```

## FillNa Based on a Condition

```python
ufo.head()
```

```python
#there are 25 null values for city
ufo.City.isnull().sum()
```

```python
ufo["Colors Reported"].value_counts()
```

```python
#here is the result.if we want to fill City col then....
ufo.loc[(ufo.City.isnull()==True)&(ufo["Colors Reported"]=="RED"),:]
```

```python
#thats it
ufo.loc[(ufo.City.isnull()==True)&(ufo["Colors Reported"]=="RED"),"City"]="New_Value_Needed"
```

## Check If 2 DataFrames Equivalent or Not

```python
drinks = pd.read_csv('../input/drinks.csv')
drinks2 = pd.read_csv('../input/drinks.csv')
```

```python
#boolean result but u cant see from here if they are exact on all values.
drinks==drinks2
```

```python
#this ll help you out.
(drinks!=drinks2).sum()
#drinks.iloc[0,0]="hi" <- u can use this then use sum() again to see the change
(drinks!=drinks2).sum()
```

## Tilde(~) Operator in Pandas

```python
#filter df bt multi categories.
```

```python
drinks.head()
```

```python
#creating multi value filter.Others are removed.
drinks2[drinks.continent.isin(["Asia","Africa"])]
```

```python
#tİLDE(~) used for selecting other than filters u specify
drinks2[~drinks.continent.isin(["Asia","Africa"])]
```
