## Selection & Filtering - loc/iloc

```python
import pandas as pd
```

```python
ufo=pd.read_csv("../input/ufo.csv")
```

```python
ufo.head()
```

```python
#loc-> [row_i_want,col_i_want]
ufo.loc[0,:]
```

```python
#as u can see/notice its inclusive on both ends.
ufo.loc[[0,1,3],:]
```

```python
ufo.loc[2:5,:]
```

```python
#DONT code like this.pandas understand that u want to get all cols but this is an implicit use and its better to  use explicit for someone
#reading yr code later. ufo.loc[0:2,:] is the one
ufo.loc[0:2]
```

```python
#selecting col u want using colname
ufo.loc[:,"City"]
```

```python
#multiple col selection
ufo.loc[:,["City","State"]]
```

```python
#slicing/using sequence for col selection
ufo.loc[:,"City":"State"]
```

```python
ufo.loc[0:2,"City":"State"]
```

```python
#here is an alternative and long way to select same thing-using drop method.axis=1 for col drop axis=0 for row drop
ufo.head(3).drop("Time",axis=1)
```

```python
#there are many ways to accomplish same thing
ufo[ufo.City=="Oakland"].head()
```

```python
#what rows do i want?-ones that have city=oakland and i want all cols
ufo.loc[ufo.City=="Oakland",:]
```

```python
#or maybe u want rows that has city=oakland and "state" cols.now we get a series.
ufo.loc[ufo.City=="Oakland","State"]
```

```python
#this could cause problems under certain scenarios.above one is better.
ufo.loc[ufo.City=="Oakland"].State
```

```python
#iloc uses integer position (int indexes in row and colwise) to choose
```

```python
#right end of iloc is exclusive.different then loc.Important to remember/note
ufo.iloc[:,0:4]
```

```python
ufo.iloc[0:3,:]
```

```python
drinks=pd.read_csv("../input/drinks.csv",index_col="country")
drinks.head()
```

## Dummy Variables

```python
#dummy variable used in statistics,econometrics and machine learning fields.
train=pd.read_csv("../input/titanic_train.csv")
```

```python
#we want to create dummy variable for the sex column-its object col now(string)
train.head()
```

```python
#creating new col that has female=0 and male=1
train["Sex_male"]=train.Sex.map({"female":0,"male":1})
```

```python
#this is a way to create dummy variables.but there is another alternative way.lets see it.
train.head()
```

```python
#this is a dataframe.get_dummies creates k cols for k catg data with 0 and 1 as values. but k-1 cols should be enough.so...
pd.get_dummies(train.Sex)
```

```python
#this is a series.Its important to note. because of the iloc indexes.if u want to create df->iloc[:,1:].same result but its a df.
pd.get_dummies(train.Sex).iloc[:,1]
```

```python
#we need to add this dataframe to our dataset so prefix to define what is this col about should be fine.
pd.get_dummies(train.Sex,prefix="Sex").iloc[:,1:]
```

```python
#lets try get_dummies for +2 catg data
train.Embarked.value_counts()
```

```python
#lets store this bcs we ll add it to our df
embarked_dummies=pd.get_dummies(train.Embarked,prefix="Embarked").iloc[:,1:]
embarked_dummies.head()
```

```python
#to make columnbind or rowbind we use concat
train=pd.concat([train,embarked_dummies],axis=1)
```

```python
train.head()
```

```python
#bonus tips-much more faster version to create dummies
import pandas as pd
train=pd.read_csv("../input/titanic_train.csv")
pd.get_dummies(train,columns=["Sex","Embarked"])
#by default sex and embarked cols are dropped.however for k catg data k-1 cols should be fine rule does not applied.
#lets solve it
```
