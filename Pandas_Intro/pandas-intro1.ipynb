## Renaming Columns in Pandas DataFrame

```python
import pandas as pd
```

```python
ufo=pd.read_csv("../input/ufo.csv")
```

```python
ufo.head()
```

```python
ufo.columns
```

```python
#solution 1 create a list that has old colnames as key new colnames as values. inplace=T do you want changes take place in yr original dataset?
#rename() is a powerful tools,creating colnames with space btw their word will cause problems when you want to call them. Dont name them w spaces.
ufo.rename(columns={"Colors Reported":"ColorsReported","Shape Reported":"ShapeReported"},inplace=True)
```

```python
ufo.columns
```

```python
#solution 2-put new names in order,doesnt need to specify which names belongs to which old colnames
ufo_cols=["city","colors_reported","shape_reported","state","time"]
```

```python
ufo.columns=ufo_cols
```

```python
ufo.columns
```

```python
#solution 3 when u are importing the file.
#header 0 means that row 0 have headers but we could override them if  we specify names arg
ufo=pd.read_csv("../input/ufo.csv",
               header=0,names=ufo_cols)
ufo.head()
```

```python
#bonus tip
#let's assume that u want to replace spaces in your cols with "_"(underscore) u could do it via rename()
#ufo.rename(columns={" ":"_"},inplace=True) let's try it out
ufo=pd.read_csv("../input/ufo.csv")
ufo.rename(columns={" ":"_"},inplace=True)
ufo.head()
#doesnt work, lets try another method...
```

```python
#string(str) attribute and replace method.
ufo.columns=ufo.columns.str.replace(" ","_")
ufo.head()
#this works
```

## Removing Columns from Pandas DataFrame

```python
#pd df is primary unit of data storage and we want to remove cols from it.
import pandas as pd
ufo=pd.read_csv("../input/ufo.csv")
```

```python
ufo.head()
```

```python
#bcs shape is dataframe attribute, u dont need to specify any args/parantheses
ufo.shape
```

```python
#solution1-drop method (its a dataframe method)(axis=0 row-axis=1 col)(we want to drop col,not row)
ufo.drop("Colors Reported",axis=1,inplace=False)
ufo.head()
#if u dont specify inplace as True then u should store this in new object otherwise yr old dataset wont be changed.
```

```python
ufo.drop("Colors Reported",axis=1,inplace=True)
ufo.head()
```

```python
#what about multiple col drop
ufo=pd.read_csv("../input/ufo.csv")
ufo.drop(["City","State"],axis=1,inplace=True)
ufo.head()
```

```python
#BONUS TIP-WHAT ABOUT IF WE WANT TO DROP ROW
#ROW names are called index/labels u need to specify indexes to remove
ufo=pd.read_csv("../input/ufo.csv")
ufo.drop([0,1],axis=0,inplace=True)
ufo.head()
#axis=0 is default u dont have to specify it
```

## Sorting in Pandas

```python
import pandas as pd
```

```python
movies=pd.read_csv("../input/imdb_1000.csv")
```

```python
movies.head()
```

```python
#its a dataframe
type(movies)
```

```python
#its a dataframe too.
type(movies.head())
```

```python
#this is a method.
type(movies.head)
```

```python
#series=vectors in R
type(movies.title)
```

```python
#sorting is changed after pandas v0.17 we are testing the new approach, not the old one before v0.17
movies.title.sort_values()
#or as an alternative if u dont want to use . notation
movies["title"].sort_values()
```

```python
#lets look its type.sort_values is  series method for this ex however we can define "by" argument so that we could use it with dataframes as well.
type(movies["title"].sort_values())
```

```python
movies["title"].sort_values(ascending=False).head()
```

```python
#up to this point we are sorting a series.now its time for sorting a dataframe.
#when u are sorting a series u get one col as an output and its string or numerical col.but when u sorting a df u get
#all types of dtypes in a col as an output so if u use type() u can see the df as a result.
```

```python
#u can add ascending=False to change it if u want
movies.sort_values("title").head()
```

```python
#bonus tip, sort by multiple arg
#df first sorted by content rating then by duration.first by approved and duration.. then by G and duration etc.
movies.sort_values(["content_rating","duration"])
```

```python
movies["content_rating"].unique()
```

```python
 movies.sort_values(['content_rating', 'duration'], ascending= [True, False]).head()
```
